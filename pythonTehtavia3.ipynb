```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#from mpl_toolkits.mplot3d import Axes3D

# Määritetään signaalin parametrit
frequency = 5  # Taajuus Hz
sampling_rate = 1000  # Näytteenottotaajuus Hz
duration = 1  # Signaalin kesto sekunneissa
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)  # Aika-akseli

# Generoidaan 5 Hz:n kompleksinen signaali
signal = np.exp(1j * 2 * np.pi * frequency * t)

# Erotellaan signaalin reaali- ja imaginaariosat
real_part = np.real(signal)
imag_part = np.imag(signal)

# Piirretään 3D-kuvaaja
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# x-akseli (aika), y-akseli (reaaliosa), z-akseli (imaginaariosa)
ax.plot(t, real_part, imag_part)

# Asettelu
ax.set_xlabel('Time [s]')
ax.set_ylabel('Real Part')
ax.set_zlabel('Imaginary Part')
ax.set_title('3D Plot of 5 Hz Complex Signal')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

freq = 5
dur = np.linspace(0,1,1000, endpoint=False)
# t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)  # Aika-akseli
ampl = 1

cosine = ampl * np.cos(2 * np.pi * freq * dur)

#set cosine to view
plt.plot(dur, cosine)

# Asettelu
plt.title("Cosine Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

freq = 5
dur = np.linspace(0 , 1 , 1000)
ampl = 1

cosine = ampl * np.cos(2 * np.pi * freq * dur)
sine = ampl * np.sin(2 * np.pi * freq * dur)

plt.plot(dur,cosine)
plt.title("Cosine Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
plt.plot(dur, sine)
plt.title("Sine Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

#print("Cosine: ", cosine)
#print("Sine: ", sine)

# Yhdistetään signaalit kompleksiseksi??
complex_signal = cosine + 1j * sine

#print(complex_signal)

# Erotellaan signaalin reaali- ja imaginaariosat
real_part = np.real(complex_signal)
imag_part = np.imag(complex_signal)



# Piirretään 3D-kuvaaja
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


# x-akseli (aika), y-akseli (reaaliosa), z-akseli (imaginaariosa)
ax.plot(dur, real_part, imag_part)

# Asettelu
ax.set_xlabel('Time [s]')
ax.set_ylabel('Cosine')
ax.set_zlabel('Sine')
ax.set_title('3D plot of cosine and sine signals')
```
