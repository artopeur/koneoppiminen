```python

'''
TehtÃ¤vÃ¤ 3


Alla olevan esimerkkikoodin avulla saat luettua world-data-2023.csv tiedostosta
195 riviÃ¤ ja 35 saraketta tietoa eri maiden tilanteesta. TehtÃ¤vÃ¤t ovat seuraavat:

1. Muokkaa Pandas data framea siten, ettÃ¤ valitset sieltÃ¤ mielenkiintoisimmat sarakkeet,
   jotka ovat:
   Country,Land Area(Km2), Armed Forces size, Co2-Emissions,Population
   Vinkki: Kysy chatGPT:ltÃ¤ neuvoa, tehtÃ¤vÃ¤n eri vaiheissa. SieltÃ¤ saat hyviÃ¤ toimivia esimerkkikoodeja kÃ¤yttÃ¶Ã¶si.

2. Taulukossa on NaN lukuja, jotka ovat ongelmiallisia maksimi arvoja haettaessa. Muuta kaikki Nan => 0.
   Taulukossa on myÃ¶s $ ja % merkkejÃ¤, jotka kannattaa muuttaa tyhjiksi merkeiksi ennen laskuoperaatioita.
   Taulukon numerot on esitetty muodossa 3,003,000 eli pilkut pitÃ¤Ã¤ muuttaa tyhjiksi merkeiksi.

3. Laske/selvitÃ¤ sen jÃ¤lkeen taulukosta
   a) MillÃ¤ maalla on suurimmat asejoukot
   b) MillÃ¤ maalla on eniten sotilaita maan kokoon verrattuna.
   c) MillÃ¤ maalla on eniten sotilaita maan vÃ¤estÃ¶Ã¶n verrattuna.
   d) Luettele 5 suurinta Co2 pÃ¤Ã¤stÃ¤jÃ¤Ã¤


'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import data_table
from vega_datasets import data


# Ota kÃ¤yttÃ¶Ã¶n Google Colabin interaktiivinen taulukkonÃ¤kymÃ¤
data_table.enable_dataframe_formatter()

#df = pd.read_csv('world-data-2023.csv')


halututSarakkeet = ['Country','Land Area(Km2)', 'Armed Forces size', 'Co2-Emissions','Population']
# Luetaan vain tietyt sarakkeet CSV-tiedostosta
df = pd.read_csv('world-data-2023.csv', usecols=halututSarakkeet)

muutettavat=['Land Area(Km2)', 'Armed Forces size', 'Co2-Emissions','Population']

# Poistetaan $ ja % merkit, sekÃ¤ pilkut
df.replace({'\$': '', '%': '', ',': ''}, regex=True, inplace=True)
df.fillna(0, inplace=True)
#Numeroiksi


df[muutettavat] = df[muutettavat].apply(pd.to_numeric, errors = 'coerce')
df['Soldiers_per_Population'] = df['Armed Forces size'] / df['Population']
df['Soldiers_per_Landmass'] = df['Armed Forces size'] / df['Land Area(Km2)']

df.fillna(0, inplace=True)

#Eniten sotilaita

# Suurimmat joukot
biggest_army = df.loc[df['Armed Forces size'].idxmax(), ['Country', 'Armed Forces size']]
print(f"Maa, jolla on suurimmat asejoukot: \n {biggest_army}")

# Suurimmat joukot maan kokoon nÃ¤hden
soldiers_per_landmass = df.loc[df['Soldiers_per_Landmass'].idxmax(), ['Country', 'Soldiers_per_Landmass']]

print(f"\nMaa, jolla on suurimmat asejoukot maan kokoon nÃ¤hden: \n {soldiers_per_landmass}")

# VÃ¤kilukuun nÃ¤hden
largest_military = df.loc[df['Soldiers_per_Population'].idxmax(), ['Country', 'Soldiers_per_Population']]
print(f"\nMaa, jolla on suurimmat asejoukot vÃ¤kimÃ¤Ã¤rÃ¤Ã¤n nÃ¤hden: \n {largest_military}")

# Viisi suurinta Co2 pÃ¤Ã¤stÃ¤jÃ¤Ã¤
columns = ['Country', 'Co2-Emissions']
largest_polluters = df[columns].nlargest(5, 'Co2-Emissions')
print(f"\nViisi suurinta Co2 pÃ¤Ã¤stÃ¤jÃ¤Ã¤:\n\n  {largest_polluters}")

# NÃ¤ytetÃ¤Ã¤n tulos
#print(df.head())

df







```
